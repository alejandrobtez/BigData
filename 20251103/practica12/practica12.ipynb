### Ejercicio 7

Crea una funci√≥n `encontrar_maximo` que use `*args` para recibir varios n√∫meros y retorne el mayor de todos. No uses la funci√≥n `max()` de Python, implementa tu propia l√≥gica.

````python
def encontrar_maximo(*args):
    """
    Recibe una cantidad variable de n√∫meros (*args)
    y retorna el mayor de todos, sin usar la funci√≥n max() de Python.
    """

    # Asumir que el primer n√∫mero es el m√°ximo
    # para tener algo con qu√© comparar.
    maximo_actual = args[0]
    
    # Iterar por todos los n√∫meros en la tupla 'args'.
    # (Podr√≠amos empezar desde el segundo, args[1:], pero
    # comparar el primero consigo mismo no afecta el resultado).
    for numero in args:
        
        # Paso 4: Si encontramos un n√∫mero m√°s grande,
        #         actualizamos nuestro 'maximo_actual'.
        if numero > maximo_actual:
            maximo_actual = numero
            
    # Paso 5: Despu√©s de revisar todos los n√∫meros, devolvemos
    #         el valor m√°s grande que encontramos.
    return maximo_actual

# --- Ejemplos de uso ---

print(f"El m√°ximo de (1, 5, 3, 9, 2) es: {encontrar_maximo(1, 5, 3, 9, 2)}")
print(f"El m√°ximo de (100, 20, 50) es: {encontrar_maximo(100, 20, 50)}")
print(f"El m√°ximo de (42) es: {encontrar_maximo(42)}")
print(f"El m√°ximo de (-5, -2, -10) es: {encontrar_maximo(-5, -2, -10)}")
````

### Ejercicio 8

Define una funci√≥n `mostrar_datos_producto` que use `**kwargs` para recibir informaci√≥n sobre un producto (nombre, precio, stock, categor√≠a, etc.) e imprima cada dato en una l√≠nea separada con formato: "clave: valor".

```python
# Ejemplo de llamada:
# mostrar_datos_producto(nombre='Laptop', precio=899.99, stock=15, marca='Dell')

```

````python
def mostrar_datos_producto(**kwargs):
    """
    Recibe una cantidad variable de argumentos por palabra clave (**kwargs)
    sobre un producto y los imprime en formato 'clave: valor'.
    """
    
    # **kwargs empaqueta todos los argumentos por palabra clave
    # en un diccionario.
    # Ej: {'nombre': 'Laptop', 'precio': 899.99, ...}
    
    print("--- Detalles del Producto ---")
    
    # Iteramos sobre el diccionario usando .items(),
    # que nos da la clave y el valor en cada paso.
    for clave, valor in kwargs.items():
        print(f"{clave}: {valor}")
    
    print("-----------------------------")

# --- Ejemplos de uso ---

# Ejemplo 1 (el solicitado)
mostrar_datos_producto(nombre='Laptop', precio=899.99, stock=15, marca='Dell')

# Ejemplo 2 (con diferentes datos)
mostrar_datos_producto(nombre='Teclado Mec√°nico', precio=120.50, categoria='Perif√©ricos')
````

### Ejercicio 10

Escribe una funci√≥n `procesar_pedido` que reciba:

- Un argumento posicional: `numero_pedido`
- Argumentos posicionales variables: `productos` (nombres de productos)
- Dos argumentos de palabra clave obligatorios: `cliente` y `direccion`
- Argumentos de palabra clave variables: `*opciones_envio`

La funci√≥n debe imprimir toda la informaci√≥n del pedido de forma organizada.

````python
def procesar_pedido(numero_pedido, *productos, cliente, direccion, **opciones_envio):
    """
    Procesa la informaci√≥n de un pedido con un formato de
    argumentos mixto y complejo.
    
    Argumentos:
    - numero_pedido (posicional): El ID del pedido.
    - *productos (variable posicional): Tupla de productos.
    - cliente (keyword-only): Nombre del cliente (obligatorio).
    - direccion (keyword-only): Direcci√≥n (obligatorio).
    - **opciones_envio (variable keyword): Dict de opciones extra.
    """
    
    print(f"\n--- Resumen del Pedido: {numero_pedido} ---")
    
    # 1. Argumentos de palabra clave obligatorios
    print(f"Cliente: {cliente}")
    print(f"Direcci√≥n: {direccion}")
    
    # 2. Argumentos posicionales variables (*productos)
    print("\nProductos:")
    if not productos:
        print("  (Sin productos en el pedido)")
    else:
        for item in productos:
            print(f"  - {item}")
            
    # 3. Argumentos de palabra clave variables (**opciones_envio)
    print("\nOpciones de Env√≠o:")
    if not opciones_envio:
        print("  (Env√≠o est√°ndar)")
    else:
        for clave, valor in opciones_envio.items():
            print(f"  - {clave}: {valor}")
            
    print("-" * (27 + len(str(numero_pedido))))

# --- Ejemplos de uso ---

# Ejemplo 1: Con todos los argumentos
procesar_pedido(101, 
                'Laptop', 'Mouse', 'Teclado', 
                cliente='Ana Garc√≠a', 
                direccion='Calle Sol 123, Madrid',
                envio_urgente=True,
                envolver_regalo=True)

# Ejemplo 2: Pedido m√≠nimo (solo los argumentos obligatorios)
procesar_pedido(102,
                cliente='Bruno D√≠az', 
                direccion='C/ Luna 45, Valencia')

# Ejemplo 3: Con un solo producto y una opci√≥n
procesar_pedido(103,
                'Cafetera',
                cliente='Carla G√≥mez',
                direccion='Av. del Mar 8, Barcelona',
                franja_horaria='Tarde (15:00-18:00)')
````

### Ejercicio 11

Crea una funci√≥n `calcular_estadisticas` que reciba n√∫meros como `*args` y retorne un diccionario con: suma, promedio, m√≠nimo y m√°ximo de esos n√∫meros.

````python
def calcular_estadisticas(*args):
    """
    Recibe una cantidad variable de n√∫meros (*args) y retorna
    un diccionario con su suma, promedio, m√≠nimo y m√°ximo.
    """
    
    # --- Manejo de caso vac√≠o ---
    # Si 'args' es una tupla vac√≠a (no se pasaron n√∫meros),
    # no podemos calcular min, max o promedio.
    if not args:
        return {
            'suma': 0,
            'promedio': None,  # Promedio de nada es indefinido
            'minimo': None,
            'maximo': None
        }
        
    # --- C√°lculos ---
    
    # Usamos las funciones integradas de Python
    suma_total = sum(args)
    cantidad = len(args)
    promedio = suma_total / cantidad
    minimo = min(args)
    maximo = max(args)
    
    # --- Creaci√≥n del diccionario de resultados ---
    resultado = {
        'suma': suma_total,
        'promedio': promedio,
        'minimo': minimo,
        'maximo': maximo
    }
    
    return resultado

# --- Ejemplos de uso ---

# Ejemplo 1
stats1 = calcular_estadisticas(1, 2, 3, 4, 5)
print(f"Estad√≠sticas de (1, 2, 3, 4, 5):\n{stats1}")

print("-" * 20)

# Ejemplo 2
stats2 = calcular_estadisticas(10, 20, 30, 40, 50, 60)
print(f"Estad√≠sticas de (10, 20, 30, 40, 50, 60):\n{stats2}")

print("-" * 20)

# Ejemplo 3 (con decimales)
stats3 = calcular_estadisticas(10.5, 5.5, 20.0)
print(f"Estad√≠sticas de (10.5, 5.5, 20.0):\n{stats3}")

print("-" * 20)

# Ejemplo 4 (caso vac√≠o)
stats4 = calcular_estadisticas()
print(f"Estad√≠sticas sin n√∫meros:\n{stats4}")
````

### Ejercicio 13

Crea una funci√≥n `calcular_descuento` que reciba `precio` (sin valor por defecto) y `porcentaje_descuento` (valor por defecto: 10). La funci√≥n debe retornar el precio final despu√©s del descuento.

````python
def calcular_descuento(precio, porcentaje_descuento=10):
    """
    Calcula el precio final tras aplicar un descuento.
    
    Argumentos:
    precio (float or int): El precio original (obligatorio).
    porcentaje_descuento (float or int): El porcentaje a descontar
                                          (valor por defecto: 10).
    
    Retorna:
    float: El precio final despu√©s del descuento.
    """
    
    descuento_aplicar = precio * (porcentaje_descuento / 100)
    precio_final = precio - descuento_aplicar
    
    return precio_final

# --- Ejemplos de uso ---

# 1. Llamada solo con el precio (usa el 10% por defecto)
precio1 = 100
precio_final1 = calcular_descuento(precio1)
print(f"Precio original: {precio1}‚Ç¨")
print(f"Descuento: 10% (por defecto)")
print(f"Precio final: {precio_final1}‚Ç¨")
````

### Ejercicio 14

Observa y predice qu√© imprimir√° el siguiente c√≥digo ANTES de ejecutarlo. Luego ejec√∫talo y explica el resultado:

````python
contador = 0

def incrementar():
    # Usamos 'global' para indicar que modificaremos
    # la variable global, no crear una local.
    global contador 
    
    contador = contador + 1
    return contador

print(f"Valor devuelto por la funci√≥n: {incrementar()}")
print(f"Valor final del contador global: {contador}")
````

üí° Explicaci√≥n de por qu√© falla
El c√≥digo falla con un error UnboundLocalError.

Esto ocurre por el √°mbito (scope) de las variables en Python:

Cuando Python lee la funci√≥n incrementar y ve la l√≠nea contador = ..., autom√°ticamente decide que contador es una variable local (que solo existe dentro de la funci√≥n).

Luego, para ejecutar la parte derecha de la asignaci√≥n (... contador + 1), intenta leer el valor de la variable local contador.

Ah√≠ est√° el problema: Est√°s intentando leer el valor de la variable local contador antes de que se le haya asignado ning√∫n valor (porque la asignaci√≥n a√∫n no ha terminado).

Python no busca la variable contador global (la que vale 0) porque la asignaci√≥n contador = ... ya lo "convenci√≥" de que contador es una variable local.

### Ejercicio 15

Crea un programa con:

- Una variable global `saldo = 1000`
- Una funci√≥n `depositar(cantidad)` que a√±ada dinero al saldo
- Una funci√≥n `retirar(cantidad)` que reste dinero del saldo
- Una funci√≥n `mostrar_saldo()` que imprima el saldo actual

Aseg√∫rate de usar correctamente la palabra clave `global` donde sea necesario.

````python
# 1. Variable Global
# Esta variable existe fuera de todas las funciones.
saldo = 1000

# 2. Funci√≥n para depositar
def depositar(cantidad):
    """
    A√±ade una cantidad al saldo global.
    """
    # Usamos 'global' para indicar que vamos a MODIFICAR
    # la variable 'saldo' que est√° fuera de esta funci√≥n.
    global saldo
    
    if cantidad > 0:
        saldo += cantidad  # (Esto es lo mismo que saldo = saldo + cantidad)
        print(f"Dep√≥sito exitoso. Saldo nuevo: {saldo}‚Ç¨")
    else:
        print("La cantidad a depositar debe ser positiva.")

# 3. Funci√≥n para retirar
def retirar(cantidad):
    """
    Resta una cantidad del saldo global, si hay fondos.
    """
    # Tambi√©n necesitamos 'global' porque vamos a MODIFICAR el saldo.
    global saldo
    
    if cantidad <= 0:
        print("La cantidad a retirar debe ser positiva.")
    elif cantidad > saldo:
        print(f"Error: Fondos insuficientes. Solo tienes {saldo}‚Ç¨.")
    else:
        saldo -= cantidad # (Esto es lo mismo que saldo = saldo - cantidad)
        print(f"Retiro exitoso. Saldo restante: {saldo}‚Ç¨")

# 4. Funci√≥n para mostrar el saldo
def mostrar_saldo():
    """
    Imprime el valor actual del saldo global.
    """
    # NOTA: Para SOLO LEER una variable global, la palabra
    # 'global' no es necesaria. Python la buscar√°
    # autom√°ticamente fuera de la funci√≥n si no la encuentra dentro.
    print(f"El saldo actual de la cuenta es: {saldo}‚Ç¨")

# --- Programa Principal (Ejemplos de uso) ---

print("--- Bienvenido al Cajero Autom√°tico ---")
# 1. Mostramos el saldo inicial
mostrar_saldo()  # Imprime 1000

print("-" * 20)

# 2. Hacemos un dep√≥sito
depositar(250)   # Modifica el global, imprime 1250

print("-" * 20)

# 3. Intentamos retirar m√°s de lo que tenemos
retirar(2000)    # Falla, el saldo sigue en 1250

print("-" * 20)

# 4. Hacemos un retiro v√°lido
retirar(400)     # Modifica el global, imprime 850

print("-" * 20)

# 5. Mostramos el saldo final
mostrar_saldo()  # Imprime 850
````

````python
# --- 1. Variable Global ---
saldo = 1000

# --- 2. Definici√≥n de Funciones ---

def depositar(cantidad):
    """
    A√±ade una cantidad al saldo global.
    """
    global saldo
    
    if cantidad > 0:
        saldo += cantidad  # (saldo = saldo + cantidad)
        print(f"Dep√≥sito exitoso. Saldo nuevo: {saldo}‚Ç¨")
    else:
        print("La cantidad a depositar debe ser positiva.")

def retirar(cantidad):
    """
    Resta una cantidad del saldo global, si hay fondos.
    """
    global saldo
    
    if cantidad <= 0:
        print("La cantidad a retirar debe ser positiva.")
    elif cantidad > saldo:
        print(f"Error: Fondos insuficientes. Solo tienes {saldo}‚Ç¨.")
    else:
        saldo -= cantidad # (saldo = saldo - cantidad)
        print(f"Retiro exitoso. Saldo restante: {saldo}‚Ç¨")

def mostrar_saldo():
    """
    Imprime el valor actual del saldo global.
    """
    # (No necesita 'global' porque solo estamos leyendo)
    print(f"\nEl saldo actual de su cuenta es: {saldo}‚Ç¨\n")

# --- 3. Funci√≥n Principal del Men√∫ ---

def menu_principal():
    """
    Muestra el men√∫ de opciones y maneja la l√≥gica
    para llamar a las otras funciones.
    """
    print("--- Bienvenido al Cajero Autom√°tico ---")
    
    # Un bucle 'while True' se ejecutar√° para siempre
    # hasta que el usuario decida 'salir' (con 'break').
    while True:
        
        # Mostramos el men√∫ en cada iteraci√≥n
        print("--- Men√∫ de Opciones ---")
        print("1. Depositar dinero")
        print("2. Retirar dinero")
        print("3. Mostrar saldo")
        print("4. Salir")
        
        # Pedimos la opci√≥n al usuario
        # .strip() elimina espacios en blanco al inicio o final
        opcion = input("Seleccione una opci√≥n (1-4): ").strip()
        
        # --- L√≥gica de decisi√≥n (el "switch-case") ---
        
        if opcion == '1':
            try:
                # Pedimos la cantidad y la convertimos a float (para decimales)
                cantidad = float(input("Introduce la cantidad a depositar: "))
                depositar(cantidad)
            except ValueError:
                print("Error: Introduce un n√∫mero v√°lido.")
        
        elif opcion == '2':
            try:
                cantidad = float(input("Introduce la cantidad a retirar: "))
                retirar(cantidad)
            except ValueError:
                print("Error: Introduce un n√∫mero v√°lido.")
        
        elif opcion == '3':
            mostrar_saldo()
            
        elif opcion == '4':
            print("Gracias por usar el cajero. ¬°Hasta luego!")
            break  # Esta es la √∫nica forma de salir del bucle 'while True'
            
        else:
            print("Opci√≥n no v√°lida. Por favor, intente de nuevo.")

# --- 4. Ejecutar el programa ---
# Llamamos a la funci√≥n del men√∫ para que todo comience.
menu_principal()
````

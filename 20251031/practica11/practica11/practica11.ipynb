Escribe un programa para recibir tres nÃºmeros enteros desde el teclado y obtener su suma y producto, calculados a travÃ©s de una funciÃ³n definida por el usuario `cal_sum_prod()`.

````python
def cal_sum_prod(n1, n2, n3):
    # Calcula la suma de los tres nÃºmeros
    suma = n1 + n2 + n3
    # Calcula el producto (multiplicaciÃ³n) de los tres nÃºmeros
    producto = n1 * n2 * n3

    return suma, producto

print("Introduce tres nÃºmeros enteros.")

# 2. Obtener la entrada del usuario
try:
    # Pedimos los tres nÃºmeros y los convertimos a entero (int)
    num1 = int(input("Introduce el primer nÃºmero: "))
    num2 = int(input("Introduce el segundo nÃºmero: "))
    num3 = int(input("Introduce el tercer nÃºmero: "))

    # 3. Llamar a la funciÃ³n y guardar los resultados
    
    # AquÃ­ "llamamos" a la funciÃ³n que definimos arriba.
    # Le pasamos num1, num2 y num3 como "argumentos".
    # La funciÃ³n nos devuelve dos valores, que guardamos en dos
    # variables nuevas: 'resultado_suma' y 'resultado_producto'.
    resultado_suma, resultado_producto = cal_sum_prod(num1, num2, num3)

    print(f"Los nÃºmeros introducidos fueron: {num1}, {num2}, y {num3}")
    print(f"La SUMA de los tres nÃºmeros es: {resultado_suma}")
    print(f"El PRODUCTO de los tres nÃºmeros es: {resultado_producto}")

except ValueError:
    print("\nError: Â¡Debes introducir solo nÃºmeros enteros vÃ¡lidos!")
````

### ğŸ§  Ejercicio 2

Un pangrama es una oraciÃ³n que utiliza todas las letras del alfabeto.

Escribe un programa que verifique si una cadena dada es un pangrama o no, utilizando una funciÃ³n definida por el usuario `ispangram()`.

````python
def ispangram(cadena):
    """
    Comprueba si una cadena de texto es un pangrama (contiene todas
    las letras del alfabeto) sin usar mÃ³dulos externos.
    Devuelve True si lo es, y False si no lo es.
    """
    
    alfabeto = "abcdefghijklmnopqrstuvwxyz"  
    # Normalizamos la cadena de entrada a minÃºsculas
    cadena_normalizada = cadena.lower()
    
    # Iteramos por CADA letra de nuestro alfabeto
    for letra in alfabeto:
        
        if letra not in cadena_normalizada:
            # Si falta UNA SOLA letra, devolvemos False
            return False            
    # Si el bucle 'for' termina (comprobÃ³ las 26 letras
    # y ninguna faltÃ³), devolvemos True.
    return True
# --- Inicio del Programa Principal ---
frase_usuario = input("Introduce una frase para comprobar: ")

if ispangram(frase_usuario):
    print(f"\n-> Â¡La frase SÃ es un pangrama!")
else:
    print(f"\n-> La frase NO es un pangrama.")
````

### ğŸ§  Ejercicio 3

Escribe un programa en Python que acepte una secuencia de palabras separadas por guiones como entrada y llame a una funciÃ³n `convert()` que la convierta en una secuencia separada por guiones despuÃ©s de ordenar las palabras alfabÃ©ticamente.

Por ejemplo, si la cadena de entrada es:

```
'here-come-the-dots-followed-by-dashes'

```

Entonces, la cadena convertida debe ser:

```
'by-come-dashes-dots-followed-here-the'

```

````python
def convert(cadena_con_guiones):
    """
    Recibe una cadena de palabras separadas por guiones.
    Devuelve una nueva cadena con las mismas palabras,
    pero ordenadas alfabÃ©ticamente y unidas por guiones.
    """
    
    # Paso 1: Dividir la cadena en una lista de palabras.
    # 'split('-')' usa el guion como separador.
    # 'here-come-the' -> ['here', 'come', 'the']
    palabras = cadena_con_guiones.split('-')
    
    # Paso 2: Ordenar la lista de palabras alfabÃ©ticamente.
    # .sort() modifica la lista 'palabras' en su lugar.
    # ['here', 'come', 'the'] -> ['come', 'here', 'the']
    palabras.sort()
    
    # Paso 3: Unir las palabras de la lista ordenada en una nueva cadena.
    # '-'.join(lista) usa el guion para "pegar" las palabras.
    # ['come', 'here', 'the'] -> 'come-here-the'
    cadena_ordenada = '-'.join(palabras)
    
    # Paso 4: Devolver el resultado final
    return cadena_ordenada

# --- Inicio del Programa Principal ---
entrada_usuario = input("Introduce tu secuencia: ")

resultado_final = convert(entrada_usuario)

print("\n--- Resultados ---")
print(f"Cadena Original:  {entrada_usuario}")
print(f"Cadena Ordenada: {resultado_final}")
````

### ğŸ§  Ejercicio 4

Escribe una funciÃ³n en Python para crear y devolver una lista que contenga tuplas de la forma

`(x, xÂ², xÂ³)` para todos los valores de `x` entre 1 y 20 (ambos incluidos).

````python
# 1. DefiniciÃ³n de la funciÃ³n

def generar_lista_tuplas():
    """
    Crea y devuelve una lista de tuplas (x, x^2, x^3)
    para x en el rango de 1 a 20 (incluidos).
    """
    # Creamos una lista vacÃ­a que funcionarÃ¡ como "acumulador"
    lista_resultados = []
    
    # 2. Iteramos en el rango
    # Usamos range(1, 21) porque el valor final (21)
    # NO se incluye, por lo que el bucle se detiene en 20.
    for x in range(1, 21):
        
        # 3. Calculamos el cuadrado y el cubo
        x_cuadrado = x ** 2
        x_cubo = x ** 3
        
        # 4. Creamos la tupla
        # (Los parÃ©ntesis son opcionales aquÃ­, pero lo hacen mÃ¡s claro)
        tupla = (x, x_cuadrado, x_cubo)
        
        # 5. AÃ±adimos la tupla a nuestra lista
        lista_resultados.append(tupla)
        
    # 6. Devolvemos la lista completa
    return lista_resultados

# --- Inicio del Programa Principal ---

# Llamamos a la funciÃ³n y guardamos la lista que devuelve
mi_lista_de_tuplas = generar_lista_tuplas()

# Imprimimos el resultado para verificar
print("Lista de tuplas generada:")
# (Usamos un bucle for para imprimirlo de forma ordenada)
for item in mi_lista_de_tuplas:
    print(item)
````

Un palÃ­ndromo es una palabra o frase que se lee igual en ambas direcciones.

Escribe un programa que defina una funciÃ³n `ispalindrome()` que verifique si una cadena dada es un palÃ­ndromo o no. Ignora espacios y diferencias entre mayÃºsculas y minÃºsculas al verificar si es palÃ­ndromo.

````python
# 1. DefiniciÃ³n de la funciÃ³n ispalindrome()

def ispalindrome(cadena):
    """
    Verifica si una cadena es un palÃ­ndromo.
    Ignora espacios y diferencias de mayÃºsculas/minÃºsculas.
    Devuelve True si es un palÃ­ndromo, False si no lo es.
    """
    
    # --- Paso 1: Normalizar la cadena ---
    
    # 1a. Convertir todo a minÃºsculas
    cadena_limpia = cadena.lower()
    
    # 1b. Quitar todos los espacios
    #     "rats live on no evil star" -> "ratsliveonnoevilstar"
    cadena_limpia = cadena_limpia.replace(" ", "")
    
    # --- Paso 2: Obtener la cadena invertida ---
    # Usamos la tÃ©cnica de "slicing" [::-1] para invertir la cadena.
    # "ratsliveonnoevilstar" -> "ratsevilonevilsitar"
    cadena_invertida = cadena_limpia[::-1]
    
    # --- Paso 3: Comparar y devolver el resultado ---
    
    # Comparamos si la versiÃ³n limpia y la invertida son idÃ©nticas.
    # Esto devolverÃ¡ automÃ¡ticamente True o False.
    return cadena_limpia == cadena_invertida

# --- Inicio del Programa Principal ---

print("Verificador de PalÃ­ndromos")

# 2. Obtener la entrada del usuario
# (O prueba con: "Anita lava la tina")
frase_usuario = input("Introduce una frase o palabra: ")

# 3. Llamar a la funciÃ³n y mostrar el resultado
if ispalindrome(frase_usuario):
    print(f"\n-> Â¡SÃ­ es un palÃ­ndromo!")
else:
    print(f"\n-> No es un palÃ­ndromo.")
````

### ğŸ§  Ejercicio 6

Escribe un programa que defina una funciÃ³n `convert()` que reciba una cadena que contiene una secuencia de palabras separadas por espacios en blanco y devuelva una cadena despuÃ©s de eliminar todas las palabras duplicadas y ordenarlas alfanumÃ©ricamente.

Por ejemplo, una cadena pasada a `convert()` podria ser:

```python
s = "Sakhi was a singer because her mother was a singer, and Sakhi's mother was a singer because her father was a singer"
```

````python
def convert(cadena):
    """
    Recibe una cadena de palabras separadas por espacios.
    Devuelve una nueva cadena con las palabras Ãºnicas,
    ordenadas alfanumÃ©ricamente.
    """
    
    # Paso 1: Dividir la cadena en una lista de palabras.
    palabras = cadena.split(' ')
    
    # Paso 2: Obtener las palabras Ãºnicas y ordenarlas.
    #
    #   set(palabras): Crea un conjunto (un 'set') a partir de la lista.
    #                  Un conjunto, por definiciÃ³n, no puede tener
    #                  elementos duplicados. Esto elimina todas
    #                  las repeticiones automÃ¡ticamente.
    #
    #   sorted(...):   Toma el conjunto y lo convierte de nuevo
    #                  en una LISTA, ordenada alfabÃ©ticamente.
    #
    palabras_ordenadas_unicas = sorted(set(palabras))
    
    # Paso 3: Unir la lista ordenada en una sola cadena.
    #         Usamos ' ' (un espacio) como "pegamento".
    resultado = ' '.join(palabras_ordenadas_unicas)
    
    # Paso 4: Devolver la cadena final
    return resultado

# --- Inicio del Programa Principal ---

# La cadena de ejemplo
s = "Sakhi was a singer because her mother was a singer, and Sakhi's mother was a singer because her father was a singer"

print("--- Cadena Original ---")
print(s)
# 2. Llamar a la funciÃ³n
cadena_convertida = convert(s)
# 3. Mostrar el resultado
print("--- Cadena Convertida (Ãšnica y Ordenada) ---")
print(cadena_convertida)
````

### ğŸ§  Ejercicio 7

Escribe un programa que defina una funciÃ³n `count_alphabets_digits()` que acepte una cadena y calcule el nÃºmero de letras y dÃ­gitos en ella.

Debe devolver estos valores como un diccionario. Llama a esta funciÃ³n con algunas cadenas de ejemplo.

````python
def count_alphabets_digits(cadena):
    """
    Acepta una cadena, cuenta el nÃºmero de letras y dÃ­gitos en ella.
    Devuelve un diccionario con los conteos "letras" y "digitos".
    """
    
    # Inicializamos los contadores a cero
    contador_letras = 0
    contador_digitos = 0
    
    # Iteramos por cada carÃ¡cter en la cadena de entrada
    for caracter in cadena:
        
        # Comprobamos si el carÃ¡cter es una letra (a-z, A-Z)
        if caracter.isalpha():
            contador_letras += 1
        
        # Si no es letra, comprobamos si es un dÃ­gito (0-9)
        elif caracter.isdigit():
            contador_digitos += 1
        
        # Si no es ni letra ni dÃ­gito (ej: espacio, coma, '!'),
        # simplemente lo ignoramos y continuamos con el siguiente.
            
    # Creamos el diccionario que vamos a devolver
    resultado = {
        "letras": contador_letras,
        "digitos": contador_digitos
    }
    
    # Devolvemos el diccionario
    return resultado

# --- Inicio del Programa Principal
# --- Ejemplo 1 ---
cadena1 = "Python es el lenguaje numero 1"
conteo1 = count_alphabets_digits(cadena1)

print(f"Cadena: '{cadena1}'")
print(f"Resultado: {conteo1}")
print(f"  -> Total Letras: {conteo1['letras']}")
print(f"  -> Total DÃ­gitos: {conteo1['digitos']}")

# --- Ejemplo 2 ---
cadena2 = "La misiÃ³n Apolo 11 fue lanzada en 1969!!!!"
conteo2 = count_alphabets_digits(cadena2)

print(f"Cadena: '{cadena2}'")
print(f"Resultado: {conteo2}")
print(f"  -> Total Letras: {conteo2['letras']}")
print(f"  -> Total DÃ­gitos: {conteo2['digitos']}")
````

### ğŸ§  Ejercicio 8

Escribe un programa que defina una funciÃ³n llamada `frequency()` que calcule la frecuencia de las palabras presentes en una cadena que se le pase.

Las frecuencias deben devolverse en orden ordenado alfabÃ©ticamente por las palabras en la cadena.

````python

def frequency(cadena):
    """
    Calcula la frecuencia de cada palabra en una cadena.
    Ignora mayÃºsculas/minÃºsculas y puntuaciÃ³n.
    Devuelve un diccionario con las palabras (claves) ordenadas
    alfabÃ©ticamente.
    """
    
    # --- Paso 1: Limpiar y Normalizar la Cadena --- 
    # Convertimos todo a minÃºsculas
    cadena_baja = cadena.lower()
    
    # Construimos una nueva cadena temporal solo con letras y espacios.
    # Esto elimina comas, puntos, nÃºmeros, etc.
    cadena_limpia = ""
    for caracter in cadena_baja:
        if caracter.isalpha() or caracter.isspace():
            cadena_limpia += caracter
            
    # --- Paso 2: Dividir la cadena en una lista de palabras ---
    # .split() sin argumentos divide por cualquier espacio
    palabras = cadena_limpia.split()
    
    # --- Paso 3: Contar las frecuencias ---
    # Usaremos un diccionario para llevar la cuenta
    conteo_temporal = {}
    
    for palabra in palabras:
        # El mÃ©todo .get(clave, valor_por_defecto) es perfecto para esto:
        # 1. Intenta obtener el conteo actual de 'palabra'.
        # 2. Si la 'palabra' no existe, usa '0'.
        # 3. Al resultado (sea el conteo o 0), le suma 1.
        conteo_temporal[palabra] = conteo_temporal.get(palabra, 0) + 1
        
    # --- Paso 4: Crear el diccionario final ordenado ---
    
    # Obtenemos una LISTA de las claves (palabras) y la ordenamos
    palabras_ordenadas = sorted(conteo_temporal.keys())
    
    # Creamos un nuevo diccionario (que en Python 3.7+
    # recuerda el orden de inserciÃ³n)
    diccionario_final_ordenado = {}
    
    # Llenamos el nuevo diccionario en el orden alfabÃ©tico
    for palabra in palabras_ordenadas:
        diccionario_final_ordenado[palabra] = conteo_temporal[palabra]
        
    # --- Paso 5: Devolver el diccionario ordenado ---
    return diccionario_final_ordenado

# --- Inicio del Programa Principal ---

print("Calculador de Frecuencia de Palabras")
print("(Se ignoran mayÃºsculas, nÃºmeros y puntuaciÃ³n)")
print("-" * 45)

# --- Ejemplo 1 ---
s1 = "El veloz murciÃ©lago hindÃº comÃ­a feliz cardillo y kiwi."
frecuencias1 = frequency(s1)

print(f"Cadena: '{s1}'")
print("Frecuencias (ordenadas alfabÃ©ticamente):")
# Imprimimos el diccionario de forma ordenada
for palabra, conteo in frecuencias1.items():
    print(f"  - {palabra}: {conteo}")

print("-" * 45)

# --- Ejemplo 2 ---
s2 = "This is a test. This is only a test, not test 1."
frecuencias2 = frequency(s2)

print(f"Cadena: '{s2}'")
print("Frecuencias (ordenadas alfabÃ©ticamente):")
for palabra, conteo in frecuencias2.items():
    print(f"  - {palabra}: {conteo}")
print("-" * 45)
````

### ğŸ§  Ejercicio 9

Escribe un programa que defina dos funciones llamadas `create_sent1()` y `create_sent2()`. Ambas reciben las siguientes 3 listas:

```python
subjects = ['He', 'She']
verbs = ['loves', 'hates']
objects = ['TV Serials', 'Netflix']

```

Ambas funciones deben formar oraciones seleccionando elementos de estas listas y devolverlas.

````python
# --- Las listas de entrada ---
subjects = ['He', 'She']
verbs = ['loves', 'hates']
objects = ['TV Serials', 'Netflix']

def create_sent1(sub, vrb, obj):
    """
    Genera todas las oraciones posibles usando bucles for anidados.
    Recibe las listas de sujetos (sub), verbos (vrb) y objetos (obj).
    Devuelve una lista de oraciones.
    """
    
    lista_oraciones = []
    
    # Bucle 1: Itera sobre los sujetos
    for s in sub:
        
        # Bucle 2: Por CADA sujeto, itera sobre los verbos
        for v in vrb:
            
            # Bucle 3: Por CADA combinaciÃ³n de sujeto-verbo, itera sobre los objetos
            for o in obj:
                
                # Creamos la oraciÃ³n (usando f-string)
                oracion = f"{s} {v} {o}."
                
                # AÃ±adimos la oraciÃ³n completa a nuestra lista
                lista_oraciones.append(oracion)
                
    # Devolvemos la lista completa
    return lista_oraciones

# 2. DefiniciÃ³n de la funciÃ³n create_sent2() (con ComprensiÃ³n de Listas)

def create_sent2(sub, vrb, obj):
    """
    Genera todas las oraciones posibles usando una comprensiÃ³n de listas.
    Recibe las listas de sujetos (sub), verbos (vrb) y objetos (obj).
    Devuelve una lista de oraciones.
    """
    
    # Esto es el equivalente en una sola lÃ­nea a los 3 bucles 'for'
    # de la funciÃ³n create_sent1.
    return [f"{s} {v} {o}." for s in sub for v in vrb for o in obj]

# --- Inicio del Programa Principal ---

print("DemostraciÃ³n del generador de oraciones")
print(f"Sujetos: {subjects}")
print(f"Verbos: {verbs}")
print(f"Objetos: {objects}")

print("-" * 40)

# --- Llamamos a la FunciÃ³n 1 ---
print("Resultados de create_sent1() (con bucles for):")
oraciones1 = create_sent1(subjects, verbs, objects)
# Imprimimos cada oraciÃ³n de la lista
for oracion in oraciones1:
    print(f"  -> {oracion}")

print("-" * 40)

# --- Llamamos a la FunciÃ³n 2 ---
print("Resultados de create_sent2() (con comprensiÃ³n de listas):")
oraciones2 = create_sent2(subjects, verbs, objects)
for oracion in oraciones2:
    print(f"  -> {oracion}")
````

````python
def get_numero_real(prompt):
    """
    FunciÃ³n auxiliar para pedir un nÃºmero (entero o real) al usuario
    y asegurarse de que la entrada sea vÃ¡lida.
    """
    while True:
        try:
            # Usamos float() para aceptar tanto enteros (ej: 5)
            # como reales (ej: 2.5)
            return float(input(prompt))
        except ValueError:
            print("Error: Debes introducir un nÃºmero vÃ¡lido.")

def analizar_plano(x, y):
    """
    Analiza un punto (x, y) en el plano 2D y determina
    su ubicaciÃ³n (Origen, Eje o Cuadrante).
    """
    punto_str = f"({x}, {y})"
    
    # --- 1. Comprobar Origen ---
    if x == 0 and y == 0:
        print(f"El punto {punto_str} estÃ¡ en el Origen (0, 0).")
    
    # --- 2. Comprobar Ejes (si no es el origen) ---
    elif x == 0:
        print(f"El punto {punto_str} estÃ¡ sobre el Eje Y.")
    elif y == 0:
        print(f"El punto {punto_str} estÃ¡ sobre el Eje X.")
        
    # --- 3. Comprobar Cuadrantes (si no estÃ¡ en ejes ni origen) ---
    elif x > 0 and y > 0:
        print(f"El punto {punto_str} pertenece al Cuadrante I (+, +).")
    elif x < 0 and y > 0:
        print(f"El punto {punto_str} pertenece al Cuadrante II (-, +).")
    elif x < 0 and y < 0:
        print(f"El punto {punto_str} pertenece al Cuadrante III (-, -).")
    elif x > 0 and y < 0:
        print(f"El punto {punto_str} pertenece al Cuadrante IV (+, -).")

def analizar_espacio(x, y, z):
    """
    Analiza un punto (x, y, z) en el espacio 3D y determina
    su ubicaciÃ³n (Origen, Eje, Plano u Octante).
    """
    punto_str = f"({x}, {y}, {z})"
    
    # --- 1. Comprobar Origen (3 ceros) ---
    if x == 0 and y == 0 and z == 0:
        print(f"El punto {punto_str} estÃ¡ en el Origen (0, 0, 0).")
    
    # --- 2. Comprobar Ejes (2 ceros) ---
    elif x == 0 and y == 0:
        print(f"El punto {punto_str} estÃ¡ sobre el Eje Z.")
    elif x == 0 and z == 0:
        print(f"El punto {punto_str} estÃ¡ sobre el Eje Y.")
    elif y == 0 and z == 0:
        print(f"El punto {punto_str} estÃ¡ sobre el Eje X.")
        
    # --- 3. Comprobar Planos (1 cero) ---
    elif z == 0:
        print(f"El punto {punto_str} estÃ¡ en el Plano XY.")
        # Opcional: Indicar el cuadrante dentro de ese plano
        analizar_plano(x, y) # Reutilizamos la funciÃ³n 2D
    elif y == 0:
        print(f"El punto {punto_str} estÃ¡ en el Plano XZ.")
    elif x == 0:
        print(f"El punto {punto_str} estÃ¡ en el Plano YZ.")
        
    # --- 4. Comprobar Octantes (ningÃºn cero) ---
    elif z > 0: # Octantes "superiores" (I a IV)
        if x > 0 and y > 0:
            print(f"El punto {punto_str} pertenece al Octante I (+, +, +).")
        elif x < 0 and y > 0:
            print(f"El punto {punto_str} pertenece al Octante II (-, +, +).")
        elif x < 0 and y < 0:
            print(f"El punto {punto_str} pertenece al Octante III (-, -, +).")
        elif x > 0 and y < 0:
            print(f"El punto {punto_str} pertenece al Octante IV (+, -, +).")
            
    elif z < 0: # Octantes "inferiores" (V a VIII)
        if x > 0 and y > 0:
            print(f"El punto {punto_str} pertenece al Octante V (+, +, -).")
        elif x < 0 and y > 0:
            print(f"El punto {punto_str} pertenece al Octante VI (-, +, -).")
        elif x < 0 and y < 0:
            print(f"El punto {punto_str} pertenece al Octante VII (-, -, -).")
        elif x > 0 and y < 0:
            print(f"El punto {punto_str} pertenece al Octante VIII (+, -, -).")

# --- Inicio del Programa Principal ---
print("Analizador de Coordenadas (Cuadrantes y Octantes)")
print("Â¿Quieres analizar un punto en el plano (2D) o en el espacio (3D)?")

while True:
    opcion = input("Escribe '2' para 2D o '3' para 3D (o 'salir'): ").strip().lower()

    if opcion == '2':
        print("\n--- Analizando en el Plano (2D) ---")
        x = get_numero_real("Introduce la coordenada x: ")
        y = get_numero_real("Introduce la coordenada y: ")
        analizar_plano(x, y)
        print("-" * 35)
        
    elif opcion == '3':
        print("\n--- Analizando en el Espacio (3D) ---")
        x = get_numero_real("Introduce la coordenada x: ")
        y = get_numero_real("Introduce la coordenada y: ")
        z = get_numero_real("Introduce la coordenada z: ")
        analizar_espacio(x, y, z)
        print("-" * 35)
        
    elif opcion == 'salir':
        print("Â¡Hasta luego!")
        break
        
    else:
        print("OpciÃ³n no vÃ¡lida. Por favor, introduce '2', '3' o 'salir'.")
````
